```python
# 0.导入必要的库 导入数据.将数据框命名为drinks 文件名drinks
import numpy as np
import pandas as pd
drinks= pd.read_csv(r'../exercise_data/drinks.csv')
drinks
```

```python
# 1.哪个大陆(continent)平均消耗的啤酒(beer)更多
drinks.groupby(by='continent').mean().sort_values(by='beer_servings').tail(1)
```

```python
# 2.打印出每个大陆(continent)的红酒消耗(wine_servings)的描述性统计值
drinks.groupby('continent').wine_servings.describe()
# drinks.loc[drinks.continent.isnull(),'continent']='NA'
```

```python
# 3.打印出每个大陆每种酒类别的消耗平均值
drinks.groupby(by='continent').mean()
```

```python
# 4.打印出每个大陆每种酒类别的消耗中位数
drinks.groupby(by='continent').median()
```

```python
# 5.打印出每个大陆对spirit饮品消耗的平均值，最大值和最小值
temp1=drinks.groupby(by='continent').mean()
temp1['mean']=drinks.groupby(by='continent').mean()['spirit_servings']
temp1['max']=drinks.groupby(by='continent').max()['spirit_servings']
temp1['min']=drinks.groupby(by='continent').min()['spirit_servings']
temp1[['mean','max','min']]
```
