```python
# 0.导入必要的库；导入数据集将数据集命名为crime，文件为：
import numpy as np
import pandas as pd
crime = pd.read_csv(r'../exercise_data/US_Crime_Rates_1960_2014.csv')
crime.head(10)
```

```python
crime['cr']=crime.Violent/crime.Population
crime.sort_values(by=['cr'])
crime.iloc[0:1]
 #1.每一列(column)的数据类型是什么样的？
crime.dtypes
```

```python
df = pd.DataFrame({'old_date': [20180501, 20181230, 20181001]})
df['new_date'] = pd.to_datetime(df['old_date'].astype(str), format='%Y%m%d')
print(df.dtypes)
# https://www.jianshu.com/p/93734eeed9b3 资料 
# 2.将Year的数据类型转换为 datetime64

def convertDatetime64(arg):
    return str(arg)+'-1'+'-1'
crime['Year1'] =pd.to_datetime(crime['Year'].apply(convertDatetime64), format='%Y-%m-%d')
crime['Year1'].dt.time
```

```python
# 3.将列Year设置为数据框的索引 index
# pd.melt(crime,id_vars=['Year'])
crime.set_index(crime['Year'],drop = True).head(10)

# crime = crime.set_index('Year', drop = True)
```

```python
# 4.删除名为Total的列
del crime['Total']
crime.head(10)
```

```python
#  5.何时是美国历史上生存最危险的年代
# 由谋杀率和犯罪率看出  
crime['cr']=crime.Violent/crime.Population
crime['mr']=crime.Murder/crime.Population
crime.sort_values(by=['mr','cr'],ascending=False)
```
