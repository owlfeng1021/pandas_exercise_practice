```python
import numpy as np
import pandas as pd
# 4.创建数据框的列名称
path='../exercise_data/iris.csv'
iris=pd.read_csv(path)
iris
```

```python
# 5.数据框中有缺失值吗？
d = iris.isnull()
row = list(iris.index)
col = list(iris.columns)
b = 0
# iris.iloc[10:20,2:3] = np.nan 
for i in row:
    for j in col:
        if pd.isnull(iris.loc[i,j]):
            b = b + 1
print(b)
iris[iris.isnull().values==True]
```

```python
# 6.将列petal_length的第10到19行设置为缺失值
# for i in range(10,20):
#     for j in col:
#         iris.loc[i,j] = np.nan
# iris.isnull()
iris.iloc[10:20,2:3] = np.nan # 需要把petal_length 设置为缺失值
iris.iloc[10:20,0:]
```

```python
# 测试代码
# for i in range(10,20):
#     for j in col:
#         iris.loc[i,j] = '1.0'
iris.iloc[10:20,0:2] = np.nan
print(iris.isnull().sum()) 
iris['5.1']=iris['5.1'].fillna('1.0') # 设置5.1的缺失值为1.0
iris.isnull().sum() 
# 7.将缺失值全部替换为1.0 
iris=iris.fillna('1.0') 
```
