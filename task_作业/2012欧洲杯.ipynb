```python
# 0.导入必要的库；导入数据集将数据集命名为euro12，文件为：
import numpy as np
import pandas as pd
euro12 = pd.read_csv(r'../exercise_data/Euro2012_stats.csv')
euro12.head(10)
```

```python
# 1..只选取 Goals 这一列
print(euro12.Goals) # 选取Goals列 1 
```

```python
# 2 有多少球队参与了2012欧洲X杯
print(len(euro12)) # 多少队伍参加 2
```

```python
# 3.该数据集中一共有多少列(columns)?
print(len(euro12.columns)) # 3 
```

```python
# 4.将数据集中的列Team, Yellow Cards和Red Cards单独存为一个名叫discipline的数据框
discipline=euro12[['Team','Yellow Cards','Red Cards']]
discipline
```

```python
# 5.对数据框discipline按照先Red Cards再Yellow Cards进行排序
print(euro12.sort_values(by=["Red Cards","Yellow Cards"],ascending=False)[["Team","Red Cards","Yellow Cards"]]) # 5.对数据框discipline按照先Red Cards再Yellow Cards进行排序 

```

```python
# 6.计算每个球队拿到的黄牌数的平均值
print(euro12["Yellow Cards"].mean())# print(euro12)# 6.计算每个球队拿到的黄牌数的平均值

```

```python
# 7.找到进球数Goals超过6的球队数据
print(euro12[euro12.Goals > 6 ]) # 7.找到进球数Goals超过6的球队数据

```

```python
# 8.选取以字母G开头的球队数据
def GetGTeam(s):
    if s[:1]=='G':
        return True
    return False
euro12[euro12["Team"].apply(GetGTeam)]
```

```python
# 9.找到英格兰(England)、意大利(Italy)和俄罗斯(Russia)的射正率(Shooting Accuracy)
euro12[(euro12['Team']=="England")|(euro12['Team']=="Italy")|(euro12['Team']=="Russia")][['Team','Shooting Accuracy']]
```
