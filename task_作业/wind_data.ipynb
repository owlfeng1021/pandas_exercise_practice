```python
# 0.将数据作存储并且设置前三列为合适的索引  文件名：wind.data
import numpy as np
import pandas as pd
wind = pd.read_csv(r'../exercise_data/wind.data',
                   sep='\s+',
                   parse_dates = [[0,1,2]]) # sep='\s+'作用是： s+可以将tab和多个空格都当成一样的分隔符 而parse_dates就是把列位为 0,1,2的解析为时间类型 并且合并为一列

print(len(wind))
print(wind.head(10))


# 3.对应每一个location，一共有多少数据值缺失
print(wind.isnull().sum())
# 4.对应每一个location，一共有多少完整的数据值
print(wind.notnull().sum())
```

```python
# 1.把超过2010年的数据都删除掉
# print(len(wind))
# print(len(wind[wind['Yr_Mo_Dy'].dt.year>2012]))
wind=wind[wind['Yr_Mo_Dy'].dt.year<=2012]
wind
```

```python
# 2.将日期设为索引，注意数据类型 
print(wind.dtypes) # 在第一个代码块中就有关于
wind.set_index(wind['Yr_Mo_Dy'])
```

```python
# 3.对应每一个location，一共有多少数据值缺失
print(wind.isnull().sum())

# 4.对应每一个location，一共有多少完整的数据值
wind.notnull().sum()
```

```python
# 5..对于全体数据，计算风速的平均值
wind.mean() # 时间类型
```

```python
# 6.创建一个名为loc_stats的数据框去计算并存储每个location的风速最小值，最大值，平均值和标准差
data = {"mean": wind.mean(), "std": wind.std(), "max": wind.max(), "min": wind.min()}
index=wind.columns
loc_stats = pd.DataFrame(data=data, index=index)
loc_stats.iloc[1:,]  # iloc 和 loc
```

```python
# 7.创建一个名为day_stats的数据框去计算并存储所有location的风速最小值
data = {"min": wind.min()}
index=wind.columns
day_stats = pd.DataFrame(data=data, index=index)
day_stats.iloc[1:,]  # iloc 和 loc
```

```python
# 8.对于每一个location，计算一月份的平均风速
def check(s):
    if s.month==1:
        return True
    return False
wind[wind['Yr_Mo_Dy'].apply(check)]
wind['January of the year']=wind['Yr_Mo_Dy'].dt.year
wind.groupby(by='January of the year').mean()
```
